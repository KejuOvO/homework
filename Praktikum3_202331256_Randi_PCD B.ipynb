```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('logo-mbokdarmi.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray], [0], None, [256], [0,256])
equalized = cv2.equalizeHist(gray)
negative = 255 - gray
brighter = cv2.convertScaleAbs(gray, alpha=1, beta=50)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs[0, 0].imshow(img_rgb)
axs[0, 0].set_title("Original")
axs[0, 1].imshow(gray, cmap='gray')
axs[0, 1].set_title("Grayscale")
axs[0, 2].imshow(negative, cmap='gray')
axs[0, 2].set_title("Negative")
axs[1, 0].imshow(brighter, cmap='gray')
axs[1, 0].set_title("Brighter")
axs[1, 1].imshow(equalized, cmap='gray')
axs[1, 1].set_title("Equalized")
axs[1, 2].plot(hist)
axs[1, 2].set_title("Histogram")
for ax in axs.flat:
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
average = cv2.blur(img, (5, 5))
median = cv2.medianBlur(img, 5)
gaussian = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
fig2, axs2 = plt.subplots(1, 4, figsize=(16, 4))
axs2[0].imshow(img_rgb)
axs2[0].set_title("Original")
axs2[1].imshow(cv2.cvtColor(average, cv2.COLOR_BGR2RGB))
axs2[1].set_title("Average Filter")
axs2[2].imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
axs2[2].set_title("Median Filter")
axs2[3].imshow(cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB))
axs2[3].set_title("Gaussian Filter")
for ax in axs2:
    ax.axis('off')
plt.tight_layout()
plt.show()
```

### 202331256_Muhammad Randi Haryanda

```python
kernel_sharp = np.array([[0, -1, 0],
                         [-1, 5, -1],
                         [0, -1, 0]])

kernel_edge = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])

sharp = cv2.filter2D(img, -1, kernel_sharp)
edge = cv2.filter2D(img, -1, kernel_edge)
```

```python
fig3, axs3 = plt.subplots(1, 3, figsize=(12, 4))
axs3[0].imshow(img_rgb)
axs3[0].set_title("Original")
axs3[1].imshow(cv2.cvtColor(sharp, cv2.COLOR_BGR2RGB))
axs3[1].set_title("Sharpened")
axs3[2].imshow(cv2.cvtColor(edge, cv2.COLOR_BGR2RGB))
axs3[2].set_title("Edge Detection")
for ax in axs3:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
